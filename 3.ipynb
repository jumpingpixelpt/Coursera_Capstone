```python
import pandas as pd
import numpy as np
import csv

import folium 
import matplotlib.cm as cm
import matplotlib.colors as colors

```

```python
df = pd.read_csv('toronto.csv')
```

```python
df.head()
```

```python
df = df[df.IsToronto]
```

<H2>TORONTO HAVE 4 Borough</H2>

```python
df.groupby(["Borough"]).size()
```

```python
map = folium.Map(location=[43.654260,-79.360636], zoom_start=11)

# add markers to map
for Latitude, Longitude, Borough, Neighborhood in zip(df['Latitude'],\
df['Longitude'], df['Borough'], df['Neighbourhood']):
    label = '{}, {}'.format(Neighborhood, Borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [Latitude, Longitude],
        radius=7,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map)  
    
map
```
