```python
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv('canada.csv')
```

```python
df
```

```python
df['Latitude'] = 0
df['Longitude'] = 0
with open('Geospatial_Coordinates.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, skipinitialspace=True)
    for row in reader:
        if df['Postal Code'].str.contains(row[0]).sum().sum()  > 0:
            df['Latitude'][df[df['Postal Code'] == row[0]].index.tolist()] = row[1]
            df['Longitude'][df[df['Postal Code'] == row[0]].index.tolist()] = row[2]   
```

```python
df
```

```python
df['IsToronto'] = df['Borough'].str.contains('Toronto')
```

```python
df
```

```python
df.to_csv(r'toronto.csv', index = False)
```
